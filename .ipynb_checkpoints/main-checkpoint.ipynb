```python
df['generation fossil brown coal/lignite'] = df['generation fossil brown coal/lignite'].astype(int)
df['generation fossil coal-derived gas'] = df['generation fossil coal-derived gas'].astype(int)
df['generation fossil gas'] = df['generation fossil gas'].astype(int)
df['generation fossil hard coal'] = df['generation fossil hard coal'].astype(int)
df['generation fossil oil'] = df['generation fossil oil'].astype(int)
df['generation fossil oil shale'] = df['generation fossil oil shale'].astype(int)
df['generation fossil peat'] = df['generation fossil peat'].astype(int)
df['total load forecast'] = df['total load forecast'].astype(int)
df['total load actual'] = df['total load actual'].astype(int)
df['price day ahead'] = df['price day ahead'].astype(int)
df['price actual'] = df['price actual'].astype(int)
df['temp'] = df['temp'].astype(int)
```

# energy demand based on weather

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_absolute_error, accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
# Select the independent variables and target variables for total load forecast
X_total_load = df[['temp', 'pressure', 'humidity', 'wind_speed', 'wind_deg']]
y_total_load = df['total load forecast']

# Split the data into training and testing sets for total load forecast
X_train_total_load, X_test_total_load, y_train_total_load, y_test_total_load = train_test_split(X_total_load, y_total_load, test_size=0.2, random_state=42)

# Create and train the Multiple Linear Regression model for total load forecast
model_total_load = LinearRegression()
model_total_load.fit(X_train_total_load, y_train_total_load)

# Make predictions on the testing set for total load forecast
y_pred_total_load = model_total_load.predict(X_test_total_load)


# Select the independent variables and target variables for weather main
X_weather_main = df[['temp', 'pressure', 'humidity', 'wind_speed', 'wind_deg']]
y_weather_main = df['weather_main']

# Split the data into training and testing sets for weather main
X_train_weather_main, X_test_weather_main, y_train_weather_main, y_test_weather_main = train_test_split(X_weather_main, y_weather_main, test_size=0.2, random_state=42)

# Scale the data for weather main
scaler = StandardScaler()
X_train_weather_main_scaled = scaler.fit_transform(X_train_weather_main)
X_test_weather_main_scaled = scaler.transform(X_test_weather_main)

# Create and train the Logistic Regression model for weather main
model_weather_main = LogisticRegression(max_iter=1000, solver='sag')
model_weather_main.fit(X_train_weather_main_scaled, y_train_weather_main)

# Make predictions on the testing set for weather main
y_pred_weather_main = model_weather_main.predict(X_test_weather_main_scaled)


# Get user input for the weather attributes
temp = float(input("Enter the temperature: "))
pressure = float(input("Enter the pressure: "))
humidity = float(input("Enter the humidity: "))
wind_speed = float(input("Enter the wind speed: "))
wind_deg = float(input("Enter the wind degree: "))

# Scale the user input
user_input_scaled = scaler.transform([[temp, pressure, humidity, wind_speed, wind_deg]])

# Predict the total load forecast based on user input
prediction_total_load = model_total_load.predict(user_input_scaled)

# Predict the weather main based on user input
prediction_weather_main = model_weather_main.predict(user_input_scaled)

# Print the predicted total load forecast and weather main
print("Predicted total load forecast:", prediction_total_load[0])
print("Predicted weather main:", prediction_weather_main[0])
```

```python
# Evaluate the model for total load forecast
mae_total_load = mean_absolute_error(y_test_total_load, y_pred_total_load)
mse_total_load = mean_squared_error(y_test_total_load, y_pred_total_load)
r2_total_load = r2_score(y_test_total_load, y_pred_total_load)

print("Evaluation metrics for Total Load Forecast:")
print("Mean Absolute Error (MAE):", mae_total_load)
print("Mean Squared Error (MSE):", mse_total_load)
print("R-squared (R2):", r2_total_load)

# Evaluate the model for weather main
accuracy_weather_main = accuracy_score(y_test_weather_main, y_pred_weather_main)

print("Evaluation metric for Weather Main:")
print("Accuracy:", accuracy_weather_main)
```
